# Importando e lendo os dados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv("../../Dados/UF_AnoMesAtend_ValorTotal.csv", encoding='ISO-8859-1', sep=';', skiprows = 3, skipfooter = 12,
                   thousands = '.', decimal = ',')
dados.head()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis
```

```python
colunas_usaveis.insert(0, 'Unidade da Federação')
```

```python
dados_usaveis = dados[colunas_usaveis]
dados_usaveis.set_index('Unidade da Federação', inplace = True)
```

```python
dados_usaveis = dados_usaveis.drop('Total', 1)
dados_usaveis
```

```python
dados_usaveis = dados_usaveis.T
```

```python
# 5 Estados a serem utilizados: 35 São Paulo, 41 Paraná, 53 Distrito Federal, 26 Pernambuco, 42 Santa Catarina
selecao = dados_usaveis[['35 São Paulo', '41 Paraná', '53 Distrito Federal', '26 Pernambuco', '42 Santa Catarina']]
```

```python
selecao.head()
```

```python
import seaborn as sns
sns.color_palette("rocket")
sns.set_palette('rocket')
sns.set_style('ticks')
```

```python
import matplotlib.ticker as ticker
```

### Visualização gráfica dos dados de 5 estados

```python
ax, fig = plt.subplots(figsize=(15,8))
ax = sns.lineplot(data = selecao, palette = 'mako')
ax.set_xticks([])
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
ax.set_title('Investimentos em Procedimentos Hospitalares por Mês/Ano Segundo Federação')
ax.set_xlabel('Agosto 2007 -- Julho 2020')
ax.legend(loc='best', bbox_to_anchor=(1, 1.2));
```
