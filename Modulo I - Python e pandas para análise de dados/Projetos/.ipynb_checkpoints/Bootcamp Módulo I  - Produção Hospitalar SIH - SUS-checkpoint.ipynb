# Parte I

## Leitura e Visualização dos dados

### Dados obtidos através do site tabnet
http://tabnet.datasus.gov.br/

```python
import pandas as pd
import numpy as np
```

```python
# Devido aos dados serem governamentais, devemos usar um tipo de enconding diferente para sua leitura
# Foi Usado Skiprows para pular as 3 primeiras linhas do arquivo
# Skipfooter usado para eliminar os dados do rodapé
# thousands e decimal foram usados para classificar a leitura dos numeros.
dados = pd.read_csv("../Dados/UF_AnoMesProc_ValorTotal.csv", encoding='ISO-8859-1', sep=';', skiprows = 3, skipfooter = 12,
                   thousands = '.', decimal = ',')
dados.head()
```

```python
# Usado para formatacao do retorno dos numeros, para sempre serem com duas casas decimais
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.info()
```

```python
import matplotlib.pyplot as plt
```

```python
dados.plot(x = 'Unidade da Federação', y = '2008/Jan', kind = 'barh', figsize = (16,8))
plt.title('Valor por UF')
```

### Para melhorar o eixo X, iremos adicionar os devidos valores, ao inves de deixar a anotação cientifica

Documentação: https://matplotlib.org/3.3.1/gallery/ticks_and_spines/tick-formatters.html

```python
import matplotlib.ticker as ticker
```

```python
ax = dados.plot(x = 'Unidade da Federação', y = '2008/Jan', kind = 'barh', figsize = (13,8))
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.title('Valor por UF');
```

## Parte II

```python
dados2 = pd.read_csv("../Dados/UF_AnoMesAtend_ValorTotal.csv", encoding='ISO-8859-1', sep=';', skiprows = 3, skipfooter = 12,
                   thousands = '.', decimal = ',')
dados2.head()
```

```python
dados2
```

```python
# Verificamos que nessa base de dados, possuimos muitos valores que são considerados nulos, por isso possuimos muitos objetos
# Iremos tratar para usar apenas os valores que são validos para nós
dados2.info()
```

### Coletando valores utilizaveis e inserindo coluna Unidade da Federação como indice

```python
# Usado o Index dos dados, enviamos para lista.
colunas_utilizaveis = dados2.mean().index.tolist()
# Inserimos a tabela com a descrição da UF
colunas_utilizaveis.insert(0, 'Unidade da Federação')
colunas_utilizaveis
```

```python
# Com o slice acima feito, agora colocamos em nosso DF para selecionar os dados que queremos
dados2_Usaveis = dados2[colunas_utilizaveis]
dados2_Usaveis.set_index('Unidade da Federação', inplace = True)
dados2_Usaveis.head()
```

```python
dados2_Usaveis.loc['12 Acre']
```

```python
# usado o T para transpor / rotacionar os dados, alterando a UF para as colunas
dados2_Usaveis.T
```

```python
dados2_Usaveis.T.plot(figsize=(13,8))
```

```python
dados2_Usaveis = dados2_Usaveis.drop('Total', axis = 1)
```

```python
import matplotlib.ticker as ticker
```

```python
ax = dados2_Usaveis.T.plot(figsize=(13,8), kind = 'line')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
ax.legend(loc='best', bbox_to_anchor=(1, 1.2));
```

```python
# Usado para setar uma seed especifica, para que os resultados sejam sempre os mesmos.
np.random.seed(687942)
```

```python
dados2_Usaveis.sample(7)
```

```python
cod_regiao = dados2_Usaveis.index.str.slice(start = 0, stop = 1)
```

```python
nome_regiao = {'1':'Norte', '2':'Nordeste', '3':'Sudeste', '4':'Sul' ,'5':'Centro-Oeste'}
nome_regiao
```

```python
# Funcao Map faz uma varredura e busca por similaridades entre os indices e os valores
# Exemplo: Para cada valor em cod_regiao(1,1,1,2,2,2...) ele irá mapear o dicionário nome_regiao e adicionar o seu valor.
coluna_a_ser_inserida = cod_regiao.map(nome_regiao)
coluna_a_ser_inserida
```

```python
dados2_Usaveis['Regiao'] = coluna_a_ser_inserida
```

### Usando ordenação por estado que mais investiu em saúde

```python
# primeiro iremos inserir o total novamente.
dados2_Usaveis['Total'] = dados2_Usaveis.sum(axis = 1)
dados2_Usaveis.head()
```

```python
# para ordenar o valor, usamos a funcao sort_values
ordenacao_por_total = dados2_Usaveis.sort_values('Total', ascending = False)
# nao queremos a coluna total de fato, entao iremos remove-la novamente, queremos apenas a ordenacao
ordenacao_por_total = ordenacao_por_total.drop('Total',axis=1)
# tivemos de dropar a regiao, pois ela estava interferindo no gráfico
ordenacao_por_total = ordenacao_por_total.drop('Regiao', axis = 1)
```

```python
ordenacao_por_total.info()
```

```python
ordenacao_por_total.head(5).T.plot(figsize=(13,8))
```

```python
ordenacao_por_total.head()
```

```python
# Iremos fazer o tratamento desse outlier de são paulo
interesse = ordenacao_por_total.columns[6:]
# Como nao existe uma explicacao, para o tanto que foi investido nas datas 2007/Out 2007/Nov 2007/Dez, iremos considerar os dados a partir de 2008
ordenacao_por_total[interesse].head()
```

```python
ordenacao_por_total[interesse].head().T.plot(figsize=(13,8))
```

### Parte IV

```python
# Para fazermos um slice, como parâmetro do que queremos plotar, podemos inserir o proprio DF e as colunas que queremos.
# Definido o total / por 1 milhao para que o eixo Y mostre diretamente os milhoes investidos
ordenacao_por_total = ordenacao_por_total / 1000000
ordenacao_por_total[ordenacao_por_total.columns[53:153]].head(5).T.plot(figsize=(13,8))
```

```python
meses_filtrados = ordenacao_por_total.columns[53:155]
```
