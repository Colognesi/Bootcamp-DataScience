## Resumo

Esta análise tem como propósito realizar uma exploração nos dados adquiridos pelo site: http://tabnet.datasus.gov.br/ <br>

O foco da análise será comparar o nível de analfabetismo e a quantia de vacinas aplicadas

### Leitura dados de imunos aplicados

```python
import pandas as pd
```

```python
imunos = pd.read_csv('Dados/imunos.csv', encoding="ISO-8859-1", skiprows=3, sep=";", 
                    skipfooter=19, thousands=".", decimal=",")
imunos.head()
```

### Leitura dados de pontuação de cobertura

```python
cobertura = pd.read_csv('Dados/cobertura.csv', encoding="ISO-8859-1", skiprows=3, sep=";", 
                    skipfooter=19, thousands=".", decimal=",")
cobertura.head()
```

### Leiura dados analfabetismo

```python
analfabetismo = pd.read_csv('Dados/analfabetismo.csv', encoding="ISO-8859-1", skiprows=3, sep=";", 
                    skipfooter=2, thousands=".", decimal=",")
analfabetismo.head()
```

## Tratamento Inicial: Remoção de números de UF

### dataset: Imunos

```python
imunos.set_index('Unidade da Federação', inplace = True)
imunos.index = imunos.index.str[3:]
imunos.head()
```

### dataset: Cobertura

```python
cobertura.set_index('Unidade da Federação', inplace = True)
cobertura.index = cobertura.index.str[3:]
cobertura.head()
```

### dataset: analfabetismo

```python
analfabetismo.set_index('Unidade da Federação', inplace = True)
analfabetismo.index = analfabetismo.index.str[3:]
analfabetismo.head()
```

### Seleção dos estados: São Paulo e Piauí

```python
renome = {'Total':'total imunos'}
imunos.rename(columns = renome, inplace = True)
imunos_loc = imunos.loc[['São Paulo', 'Piauí']]['total imunos']
```

```python
renome = {'Total':'total analfabetismo'}
analfabetismo.rename(columns = renome, inplace = True)
analfabetismo_loc = analfabetismo.loc[['São Paulo', 'Piauí']]['total analfabetismo']
```

```python
renome = {'Total':'total cobertura'}
cobertura.rename(columns = renome, inplace = True)
cobertura_loc = cobertura.loc[['São Paulo', 'Piauí']]['total cobertura']
```

```python
selecao = pd.DataFrame([imunos_loc, analfabetismo_loc, cobertura_loc])
```

### Adição da população:
Esta adição esta sendo realizada, pois a minha hipótese é que estados com menores indices de analfabetismo possuem uma maior taxa de vacinação.

```python
selecao = selecao.T
selecao
```

```python
# Fonte: https://pt.wikipedia.org/wiki/S%C3%A3o_Paulo | https://pt.wikipedia.org/wiki/Piau%C3%AD
    
populacao = {'São Paulo':12252023, 'Piauí': 3264531 }
selecao['populacao'] = selecao.index.map(populacao)
```

### Realizando a contagem de imunos por população aplicados

```python
selecao['imunos por pop'] = selecao['populacao'] / selecao['total imunos']
selecao = selecao.sort_values('imunos por pop', ascending = False)
selecao
```

Apesar da diferença da taxa de analfabetismo ser alta, podemos verificar que a cobertura total de imunos aplicadas é bem próxima.<br>
A cobertura de São Paulo é maior, porém a quantia de vacinas aplicadas por população, do Piauí é maior.

### Visualização Barras:

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

sns.set_palette('Accent')
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(13,8))
ax = sns.barplot(data = selecao, x = 'imunos por pop', y = selecao.index, orient = 'h')
ax.set_title('Quantia de imunos aplicados por população')
ax.set_ylabel('Estados selecionados')
ax.set_xlabel('Imunos(porcentual) aplicados por população')
plt.xticks(rotation = 30);
```

Com o gráfico acima, podemos ver que a taxa de imunizações realizadas no estado do Piauí, foi muito superior a de São Paulo.<br>
Essa pode ser uma das causas do Piauí nao estar na lista de um dos estados mais afetados pelo surto de febre amarela, que ocorreu em 2018.<br>
Decidi com esta forma, de selecionar a quantia aplicada por população, pois a comparação direta seria injusta, em vista que a população de São Paulo é muito maior que a do estado do Piauí.<br>

### Verificando o percentual de analfabetismo e vacinação dos estados:

```python
plt.figure(1, figsize=(25,15))
plt.subplot(221)
ax = sns.barplot(data = selecao, orient = 'h', x = 'total analfabetismo', y = selecao.index)
plt.xticks(rotation=30)
ax.set_title('Percentual de analfabetismo dos estados: \n Piauí e São Paulo')
ax.set_ylabel('Estados selecionados')
ax.set_xlabel('')

plt.subplot(222)
ax = sns.barplot(data = selecao, x = 'imunos por pop', y = selecao.index, orient = 'h')
plt.xticks(rotation=30)
ax.set_title('Imunos(percentual) aplicados por população dos estados: \n Piauí e São Paulo')
ax.set_ylabel('')
ax.set_xlabel('');
```

Com os gráficos acima, lado a lado, podemos verificar as quantias em porcentagem das vacinas aplicadas, segundo a pesquisa mencionada acima e o que podemos visualizar agora, ambas as variáveis aparentam ter uma correlação. Mas isso não implica diretamente causalidade, o baixo nível de vacinas pode variar de orçamento, propaganda governamental falha, discrença da população na vacina ou na obrigatoriedade da mesma, existem diversas variáveis em questão.
