# Leitura dos dados

```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('../Dados/UF_AnoMesAtend_ValorTotal.csv', encoding="ISO-8859-1", skiprows=3, sep=";", 
                    skipfooter=12, thousands=".", decimal=",")
dados.head()
```

## Tratamento 1:
inserir a média como valor alternativo na coluna 2020/Ago para o Acre

```python
tratamento = dados.T[1]
tratamento.replace('-', np.nan, inplace = True)
tratamento.fillna(0, inplace = True)
tratamento.drop('Total', inplace =True)
```

```python
valor_a_ser_inserido = round(tratamento.tail(-1).astype('float').mean(), 2)
```

```python
dados['2020/Ago'].replace('-', valor_a_ser_inserido, inplace = True)
dados['2020/Ago'] = dados['2020/Ago'].astype('float')
dados['2020/Ago'].head()
```

## Tratamento 2: Selecionar as colunas utilizaveis

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
```

```python
dados_usaveis = dados[colunas_usaveis]
dados_usaveis.set_index('Unidade da Federação', inplace = True)
```

```python
dados_usaveis = dados_usaveis.sort_values('Total', ascending = False)
dados_usaveis.head()
```

## Tratamento 3: Removendo numeros dos indices, deixando apenas o nome do estado

```python
dados_usaveis.index = dados_usaveis.index.str[3:]
dados_usaveis.head()
```

## Tratamento 4: Dropando o total

```python
dados_usaveis = dados_usaveis.drop('Total', 1)
dados_usaveis.head()
```

### Realizando slice do ultimo mês de investimentos em procedimentos hospitalares

### Criar uma função para a limpeza e tratamento dos dados, retornando duas tabelas "população" e "unida".

```python
def limpeza_e_tratamento(var_url, nome_col_uf, nome_col_pop, nome_uf_new, nome_pop_new, coluna_desejada):
    '''
    Imports necessários: Pandas
    Váriaveis:
    var_url = uma url salva em uma variável
    nome_col_uf = nome da coluna contendo a unidade federativa
    nome_col_pop = nome da coluna contendo população
    nome_uf_new = o novo nome a ser atribuido a coluna unidade federativa
    nome_pop_new = novo nome a ser atribuido a coluna população
    coluna_desejada = coluna a ser utilizada, ou seja, um slice feito do dataframe original
    '''
    
    
    import pandas as pd
    
    # leitura HTML dos dados(dados usados foram extraidos do wikipedia)
    dados = pd.read_html(var_url, thousands = ' ')
    #selecao realizada para selecionar a primeira tabela encontrada no site.
    selecao = dados[0]
    #selecao feita para selecionar as colunas de Unidade federativa e População
    selecao = selecao[[nome_col_uf, nome_col_pop]]
    #dicionario criado para fazer a alteração dos nomes das colunas
    nomeacao_colunas = {nome_col_pop:nome_pop_new, nome_col_uf:nome_uf_new}
    #realizando a renomeação das colunas
    selecao = selecao.rename(columns = nomeacao_colunas)
    #setando index como o novo nome escolhido
    selecao = selecao.set_index(nome_uf_new)
    #parte abaixo comentada pois o tratamento dos STR foram feitos anteriormente.
    #juntar.index = juntar.index.str[3:]
    #juntar.index = juntar.index.str.strip()
    selecao.index = selecao.index.str.strip()
    #dataframe com retorno das tabelas de populacao e a coluna selecionada
    df_atualizado = selecao.join(coluna_desejada)
    
    return(selecao, df_atualizado)
    
```

### Testando a funcao:

```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o'
```

```python
coluna_desejada = round(dados_usaveis[dados_usaveis.columns[-1]])
```

```python
populacao, junto = limpeza_e_tratamento(url, 'Unidade federativa', 'População', 'uf', 'populacao', coluna_desejada)
```
