### Leitura dos dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('../Dados/UF_AnoMesAtend_ValorTotal.csv', encoding="ISO-8859-1", skiprows=3, sep=";", 
                    skipfooter=12, thousands=".", decimal=",")
dados.head()
```

### selecionando colunas utilizaveis

```python
colunas = dados.mean().index.tolist()
colunas.insert(0, 'Unidade da Federação')
```

```python
dados = dados[colunas].set_index('Unidade da Federação').sort_values('Total', ascending = False)
```

```python
dados.drop('Total', 1, inplace = True)
```

### consertar index

```python
dados.index = dados.index.str[3:]
dados.head()
```

### Selecionando colunas a partir de fevereiro 2008

```python
selecionados = dados[dados.columns[6:]]
```

### Convertendo as datas

```python
import datetime
```

```python
meses = {
    'Jan': 1,
    'Fev': 2,
    'Mar': 3,
    'Abr': 4,
    'Mai': 5,
    'Jun': 6,
    'Jul': 7,
    'Ago': 8,
    'Set': 9,
    'Out': 10,
    'Nov': 11,
    'Dez': 12,
}

def para_dia(ano_mes):
    ano: int = int(ano_mes[:4])
    mes: str = ano_mes[5:]
    mes_numero: int = meses[mes]
    return datetime.date(ano, mes_numero, 1)
```

```python
selecionados.columns = selecionados.columns.map(para_dia)
selecionados.head()
```

### Melt

```python
mensal = selecionados.T
mensal_aberto = mensal.reset_index().melt(id_vars = ['index'], value_vars = mensal.columns)
mensal_aberto.head()
```

```python
mensal_aberto.columns = ['ano_mes_dia', 'uf', 'valor']
```

```python
mensal_aberto['ano_mes_dia'] = mensal_aberto['ano_mes_dia'].astype('datetime64')
```

```python
mensal_aberto['mes'] = mensal_aberto['ano_mes_dia'].dt.month
mensal_aberto['ano'] = mensal_aberto['ano_mes_dia'].dt.year
```

# Insersao de dias mensais

```python
dias_por_mes = {
    1 : 31,
    2 : 28,
    3 : 31,
    4 : 30,
    5 : 31,
    6 : 30,
    7 : 31,
    8 : 31,
    9 : 30,
    10 : 31,
    11 : 30,
    12 : 31
}
```

```python
mensal_aberto['gasto_diario'] = mensal_aberto['valor'] / mensal_aberto['mes'].map(dias_por_mes)
```

```python
mensal_aberto
```

### Plotando gráfico

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

sns.set_style('darkgrid')
sns.set_palette('Accent')
```

```python
meses = {
    1:'Janeiro',
    2:'Fevereiro',
    3:'Março',
    4:'Abril',
    5:'Maio',
    6:'Junho',
    7:'Julho',
    8:'Agosto',
    9:'Setembro',
    10:'Outubro',
    11:'Novembro',
    12:'Dezembro'
}
```

```python
mensal_aberto['nome_mes'] = mensal_aberto['mes'].map(meses)
```

```python
mensal_aberto
```

```python
plt.figure(figsize=(13,8))
ax = sns.lineplot(data = mensal_aberto.query("uf in ['São Paulo', 'Ceará', 'Bahia', 'Paraná'] & ano == 2019"), 
                  x = 'nome_mes', y = 'gasto_diario', hue = 'uf', sort = False, palette = 'rocket')
ax.set_ylabel('Investimentos diariamente')
ax.set_title('Investimentos Diarios \n Decorrer do Ano de 2019')
ax.set_xlabel('')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[1:], labels=labels[1:], loc = "best", bbox_to_anchor=(1,1.01))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.xticks(rotation = 35)
plt.show()
```
